# Exercício Proposto: Sistema de Gerenciamento de Biblioteca

Imagine que você está construindo um sistema para gerenciar os itens de uma biblioteca. Existem diferentes tipos de itens, e o sistema precisa rastrear suas informações, status de empréstimo e algumas regras específicas.

## Requisitos do Exercício

1. Classe Base: ItemBiblioteca

- Crie uma classe ItemBiblioteca com os seguintes atributos de **instância**:
    - titulo (string)
    - autor_ou_diretor (string)
    - ano_publicacao (inteiro)
    - disponivel (booleano, padrão True)
- Métodos:
    - __init__: Construtor para inicializar os atributos.
    - apresentar_detalhes(): Um método que imprime todos os detalhes do item, incluindo seu status de disponibilidade.
    - emprestar_item(): Tenta emprestar o item. Se disponivel for True, muda para False e imprime uma mensagem de sucesso. Caso contrário, imprime uma mensagem de que o item não está disponível.
    - devolver_item(): Tenta devolver o item. Se disponivel for False, muda para True e imprime uma mensagem de sucesso. Caso contrário, imprime uma mensagem de que o item já está disponível.


````python
class ItemBiblioteca():
    def __init__(self, titular:str, autor_ou_diretor:str, ano_publicacao:int, disponivel:bool):
        self.titular = titular
        self.autor_ou_diretor = autor_ou_diretor
        self.ano_publicacao = ano_publicacao
        self.disponivel = disponivel
    
    # Criação dos getter
    @property
    def titular(self):
        return self._titular
    
    @property
    def autor_ou_diretor(self):
        return self._autor_ou_diretor

    @property
    def ano_publicacao(self):
        return self._ano_publicacao

    @property
    def disponivel(self):
        return self._disponivel

    # Criação dos setter
    @titular.setter
    def titular(self, titular):
        if not isinstance(titular, str) or not titular.strip():
            raise ValueError(f'{titular} não é um valor valido. Por favor informar um valor valido para o campo titular ou não deixe o campo em branco.')
        self._titular = titular
    
    @autor_ou_diretor.setter
    def autor_ou_diretor(self, autor_ou_diretor):
        if not isinstance(autor_ou_diretor, str) or not autor_ou_diretor.strip():
            raise ValueError(f'Por favor informar um valor valido para o {autor_ou_diretor} ou não deixo o campo em branco.')
        self._autor_ou_diretor = autor_ou_diretor

    @ano_publicacao.setter
    def ano_publicacao(self, ano_publicacao):
        if not isinstance(ano_publicacao, int) or ano_publicacao < 0:
            raise ValueError(f'O ano informado não pode conter valor negativo, texto ou número decimal. Por favor informe um valor válido.')
        self._ano_publicacao = ano_publicacao
    
    @disponivel.setter
    def disponivel(self, disponivel):
        if not isinstance(disponivel, bool):
            raise ValueError('Por favor informar um valor True ou False nesse campo.')
        self._disponivel = disponivel
    
    def __str__(self):
        """
            Método para imprimir todos os detalhes do item, 
            incluindo seu status de diponibilidade.

            Args:
                void: Sem argumentos
            Returns:
                String: Retorna uma string com os detalhes do item.
        """
        return f"""Informações Gerais:
Titular:{self._titular}
Autor/Diretor: {self._autor_ou_diretor}
Ano De Lançamento: {self._ano_publicacao}
Disponibilidade: {"Disponivel" if self._disponivel else "Não Disponivel"}
"""


    def emprestar_item(self):
        """
            Método para emprestar item se disponivel e imprimir uma
            mensagem com sucesso ou falha
            
            Args:
                void - não recebe nenhum argumento
            Returns:
                void: não retorna nenhum valor, só imprime uma mensagem
                      de sucesso ou de falha.
        """
        if self._disponivel:
            self._disponivel = False
            print(f"O Item {self._titular} disponibilizado com sucesso.")
        else:
            print(f"O Item {self._titular} não se encontra disponivel no momento.")
        print()

    def devolver_item(self):
        """
            Método criado para marcar a devolução do item
            Args:
                void - não tem nenhum argumento
            Returns: 
                void -  Não retorna nenhuma valor, só imprime uma 
                        de devolução com sucesso ou falha.
        """
        if not self._disponivel:
            self._disponivel = True
            print(f"Devolução do item {self._titular} com sucesso.")
        else:
            print(f"O item '{self._titular}' em questão já encontra disponível.")

IB1 = ItemBiblioteca("Biblioteca Central", "Rachel de Queiroz", 1930, True)
print(IB1)
IB1.devolver_item()
print(IB1)
````

2. Herança: Livro (Herda de ItemBiblioteca)

- Crie uma classe Livro que herda de ItemBiblioteca.
- Atributos de instância adicionais:
    - isbn (string)
    - genero (string)

- Métodos:
    - __init__: Sobrescreva o construtor para aceitar os novos atributos e chamar o construtor da classe pai.
    - apresentar_detalhes(): Sobrescreva este método para adicionar os detalhes específicos de Livro (ISBN, gênero) aos detalhes da ItemBiblioteca.

````python
class Livro(ItemBiblioteca):
    def __init__(self, titular, autor_ou_diretor, ano_publicacao, disponivel, isbn:str, genero:str):
        super().__init__(titular, autor_ou_diretor, ano_publicacao, disponivel)
        self.isbn = isbn
        self.genero = genero
    
    @property
    def isbn(self):
        return self._isbn

    @property
    def genero(self):
        return self._genero

    @isbn.setter
    def isbn(self, isbn):
        if not isinstance(isbn, str) or not isbn.strip():
            raise ValueError('Por favor informar um valor valido ou não deixe esse campo em branco.')
        self._isbn = isbn
    
    @genero.setter
    def genero(self, genero):
        if not isinstance(genero, str) or not genero.strip():
            raise ValueError('Por favor informar um valor valido ou não deixe esse campo em branco.')
        self._genero = genero
    
    def __str__(self):
        return f"""{super().__str__()}\nISBN: {self._isbn}\nGenero: {self._genero}
"""

LV1 = Livro("Machado de Assis", "Machado de Assis", 1899, True, "978-85-01-00206-4", "Romance")
print(LV1)
````

3. Herança Multinível: LivroDigital (Herda de Livro)

    - realizar_download(): Um método exclusivo para LivroDigital que imprime uma mensagem sobre o download do livro no formato especificado.

````python
class LivroDigital(Livro):
    TIPO_ACESSO:str = "Download Online"
    
    def __init__(self, titular, autor_ou_diretor, ano_publicacao, 
                 disponivel, isbn, genero, 
                 formato, tamanho):
        super().__init__(titular, autor_ou_diretor, ano_publicacao, 
                         disponivel, isbn, genero)
        self.formato = formato
        self.tamanho = tamanho

    @property
    def formato(self):
        return self._formato

    @formato.setter
    def formato(self, formato):
        if not isinstance(formato, str) or not formato.strip():
            raise ValueError('O formato informado não é correto ou campo formata está vazio. Por favor verificar esse campo.')
        self._formato = formato

    @property
    def tamanho(self):
        return self._tamanho
    
    @tamanho.setter
    def tamanho(self, tamanho:float = 0):
        if not isinstance(tamanho, (float, int)) or tamanho < 0:
            raise ValueError('O valor informado para o tamanho do arquivo é invalido, por favor rever isso.')
        self._tamanho  = tamanho
    
    def realizar_download(self):
        print(f'Realizando o Download do livro {self._titular} de tamanho {self._tamanho}')

    def __str__(self):
        return f"{super().__str__()}\nTamanho: {self._tamanho}\nFormato: {self._formato}"

LD1 = LivroDigital("Machado de Assis",
                   "Machado de Assis", 
                   1899, 
                   True, 
                   "978-85-01-00206-4", 
                   "Romance", 
                   "PDF",
                    2000)

print(LD1)
````

````python
class DVD(ItemBiblioteca):
    def __init__(self, titular, autor_ou_diretor, ano_publicacao, 
                 disponivel, duracao_minutos:int, idioma:str):
        super().__init__(titular, autor_ou_diretor, ano_publicacao, 
                         disponivel)
        self.duracao_minuto = duracao_minutos
        self.idioma = idioma
    
    @property
    def duracao_minuto(self):
        return self._duracao_minuto

    @duracao_minuto.setter
    def duracao_minuto(self, duracao_minuto):
        if not isinstance(duracao_minuto, (float, int)) or duracao_minuto < 0:
            raise ValueError('Informe a duração do filme corretamente.')
        self._duracao_minuto = duracao_minuto

    @property
    def idioma(self):
        return self._idioma

    @idioma.setter
    def idioma(self, idioma):
        if not isinstance(idioma, str) or not idioma.strip():
            raise ValueError('Informe o idioma do filme corretamente não pode ser um campo vazio.')
        self._idioma = idioma

    def __str__(self):
        return f"{super().__str__()}\
        \nDurcação em Minutos: {self._duracao_minuto}\
        \nIdioma: {self.idioma}"

dvd1 = DVD("Biblioteca Central", "Rachel de Queiroz", 1930, True, 200, 'Português')
print(dvd1)

````

- super().__init__(): Essencial para chamar o construtor da classe pai ao inicializar uma subclasse.

    def valor(self, novo_valor):
        if novo_valor >= 0:
            self._valor = novo_valor
